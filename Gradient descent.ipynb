```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[55, 71, 68, 87, 101, 87, 75, 78, 93, 73]
y=[91, 101, 87, 109, 129, 98, 95, 101, 104, 93]
```

```python
def leastsquares(x, y): 
    m=len(x)
    x=np.array(x).reshape(m,1) #转化为列矩阵
    y=np.array(y).reshape(m,1) #转化为列矩阵
    x_1=np.ones(m).reshape(m,1) #一个都为1的列矩阵
    X=np.append(x, x_1, axis=1) #将x与x_1合并为一个二列的矩阵
    X_T=X.transpose() #X的转置矩阵
    w=np.dot(np.dot(np.linalg.inv(np.dot(X_T,X)),X_T),y) #最小二乘法的公式
    w0=w[1]
    w1=w[0]
    return w0, w1
```

```python
def gradientdescent(x, y, iteration):
    m=len(x)
    x=np.array(x).reshape(m,1) #转化为列矩阵
    y=np.array(y).reshape(m,1) #转化为列矩阵
    x_1=np.ones(m).reshape(m,1) #一个都为1的列矩阵
    X=np.append(x_1, x, axis=1) #将x_1与x合并为一个二列的矩阵
    alpha=0.0001
    theta = np.array([1, 1]).reshape(2, 1) #初始斜率与初始截距构成的2*1的矩阵
    diff = np.dot(X, theta) - y
    gradient = (1./m) * np.dot(np.transpose(X), diff) #求梯度公式
    for i in range(iteration):
        theta = theta - alpha * gradient #更新参数
        diff = np.dot(X, theta) - y
        gradient = (1./m) * np.dot(np.transpose(X), diff) 
    w0=theta[0]
    w1=theta[1]
    return w0,w1
```

```python
w0_leastsquares, w1_leastsquares = leastsquares(x, y) 
```

```python
w0_gradientdescent, w1_gradientdescent=gradientdescent(x, y, 10000)
```

```python
plt.scatter(x,y,c='red')
x_leastsquares=np.arange(50,110,0.1)
y_leastsquares=w1_leastsquares*x_leastsquares+w0_leastsquares
plt.plot(x_leastsquares,y_leastsquares)
plt.show()
```

```python
plt.scatter(x,y,c='red')
x_gradientdescent=np.arange(50,110,0.1)
y_gradientdescent=w1_gradientdescent*x_gradientdescent+w0_gradientdescent
plt.plot(x_gradientdescent,y_gradientdescent)
plt.show()
```
